### 1. Basic String Manipulation:

Problem:
Write a Python program that:

- Accepts a list of sentences.
- For each sentence, reverses the order of words (not the characters).
- Ignores punctuation while reversing the words.
- Returns the new list of sentences.

```python
lst = ["Hello world", "Python is awesome!"]
res = []
for str in lst:
    words: list[str] = str.split()
    words = words[::-1]
    res.append(' '.join(words))
res
```

### 6. Recursion and Factorial Calculation:

Problem:
- Write a recursive Python function to calculate the factorial of a number. If the input is less than 0, return "Input must be a non-negative integer".

```python
def factorial(n):
    if n < 0:
        return "Input must be a non-negative integer."
    if n == 0 or n == 1:
        return n
    return n * factorial(n - 1)

num = 5
print(factorial(5))
```

### 9. Class and Object-Oriented Programming:

- Has a method display_details() that prints out the car's details in the following format: "Make: <make>, Model: <model>, Year: <year>".

### 2. Grouping and Counting with Dictionaries:

Problem:
- Write a Python program that:
- Accepts a list of strings representing different fruits purchased by customers.
- Groups the fruits by their type (e.g., citrus, berry, stone, etc.) and counts how many of each type were purchased.
- Return a dictionary where the keys are the fruit types and the values are the counts.

```python
fruits = ["apple", "orange", "banana", "blueberry", "apple", "orange", "cherry"]

freq = {}

for fruit in fruits:
    freq[fruit] = freq.get(fruit, 0) + 1

freq
```

```python
from typing import Counter


fruits = ["apple", "orange", "banana", "blueberry", "apple", "orange", "cherry"]

freq = Counter(fruits)
list(freq)
```

```python
fruits = ["apple", "orange", "banana", "blueberry", "apple", "orange", "cherry"]

freq = {}

for fruit in fruits:
    if fruit in freq:
        freq[fruit] += 1
    else:
        freq[fruit] = 1

freq
```

### 3. Merging and Modifying Multiple Dictionaries:

Problem:
- Write a Python program that:
- Accepts two dictionaries: one containing employee names and their respective departments, and the other containing employee names and their respective salaries.
- Merges these two dictionaries into one, where the key is the employee name and the value is a dictionary containing both the department and salary.
- Add a new key total_salary for each employee, which is the sum of the salary and an additional 10% bonus.

```python
employees = {
    "Alice": "HR",
    "Bob": "IT",
    "Charlie": "Finance"
}

salaries = {
    "Alice": 50000,
    "Bob": 60000,
    "Charlie": 70000
}

res = {}

for employee in employees.items():
    res[employee[0]] = {"department": employee[1]}

for salary in salaries.items():
    res[salary[0]]["salary"] = salary[1]
    res[salary[0]]["total_salary"] = salary[1] + round(0.1*salary[1],1)
    
res
```

### 4. Lambda, Filter, and Map Functions with Complex Data:

Problem:
- Write a Python program that:
- Accepts a list of dictionaries, each representing a product with name, category, and price.
- Use filter() to return only the products that are in the "electronics" category and cost more than 500.
- Use map() with a lambda function to increase the price of the remaining products by 10%.
- Return the updated list of products.

```python
products = [
    {"name": "Laptop", "category": "electronics", "price": 1200, },
    {"name": "Shoes", "category": "fashion", "price": 100},
    {"name": "Smartphone", "category": "electronics", "price": 800},
    {"name": "T-shirt", "category": "fashion", "price": 50}
]

def filter_func(product):
        return product["category"] == "electronics" and product["price"] > 500 

res = list(filter(filter_func, products))

# increase price by 10%
res = list(map(lambda x: {**x, "price": x["price"] + int(0.1*x["price"])}, res))
res
```
