# Data Visualization with Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
a = [1,2,3,4,5]
b = [1,4, 9, 16, 25]

# line plot
plt.plot(a,b)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Chart')
```

```python
a = [1,2,3,4,5]
b = [1,4, 9, 16, 25]

# line plot
plt.plot(a,b, color='red', linestyle='--', marker='o', linewidth=3, markersize=9)
plt.grid()
```

## Multi plots

```python
## Sample data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y1,x,color='blue')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(y1,x,color='red')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(y1,x,color='purple')
plt.title("Plot 4")
```

```python
## Bar plot
categories=['A','B','C','D','E']
values=[5,7,3,12,8]

plt.bar(categories,values, color='green')
```

```python
## Histogram

data=[1,2,3,3,3,3,3,4,4,4,4,5,6,7,7,8,8,8,8,8,9,10]

plt.hist(data, color='orange', edgecolor='black')
```

```python
## Scatter plot

x=[1, 2, 3, 4, 5, 6]
y=[2, 3, 4, 5, 6, 7]

plt.scatter(x,y, marker='x')
```

```python
### pie chart

labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0) ##move out the 1st slice

##create apie chart
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

## Sales Data Visualization

```python
import pandas as pd
sales_data_df = pd.read_csv('sales.csv')
sales_data_df=sales_data_df.head(50)
```

```python
sales_data_df.info()
```

```python
sales_by_product= sales_data_df.groupby('Item Type')['Total Revenue'].sum()
sales_by_product
```

```python
sales_by_product.plot(kind='bar', color='blue')
```

```python
channel_type = sales_data_df.groupby('Sales Channel')['Total Revenue'].sum()
channel_type
```

```python
# Create pie chart for Sales Channel distribution
channel_type.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 6), colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.ylabel('')
plt.title('Sales Revenue by Channel')
plt.show()
```
