```python
genre_url=f"https://api.themoviedb.org/3/genre/movie/list?api_key={API_KEY}&language=en-US"
```

```python
# Fetch all genre and save it in genre_df
genre_response = requests.get(genre_url)
genre_df = pd.DataFrame(genre_response.json()['genres'])
genre_df
```

```python
response = requests.get(url)
```

```python
demo_df = pd.DataFrame(response.json()['results'])[['id','title','overview','genre_ids']]
```

```python
# iterate demo_df and change genre_ids by its field from genre_df
for i in range(len(demo_df)):
  demo_df['genre_ids'][i] = [genre_df[genre_df['id'] == x]['name'].values[0] for x in demo_df['genre_ids'][i]]
```

```python
for i in range(1, 534):
      url=f"https://api.themoviedb.org/3/movie/top_rated?api_key={API_KEY}&language=en-US&page={i}"
      response = requests.get(url)
      response_json = response.json()
      if 'results' in response_json and response_json['results']:
          temp_df = pd.DataFrame(response_json['results'])[['id','title','overview','genre_ids']]
          df = pd.concat([df, temp_df], ignore_index=True)
      else:
          print(f"No results or unexpected response for page {i}. Stopping data collection.")
          # print(response_json) # Uncomment for debugging to see the problematic response
          break
```
