## Pandas

```python
import numpy as np
import pandas as pd
```

```python
## Pandas Series
data = np.array([1,2,3,4,5])
series = pd.Series(data)
print("Series \n",series)
print(type(series))
```

```python
# Series from Dictionary
data = {'a':1,'b':2,'c':3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## Dataframe
# Create dataframe from a dictionary of list 

data = {
    'Name': ['Krish', 'John', 'Jack'],
    'Age': [25, 30, 45],
    'City': ['Bangalore', 'New york', 'California']
}

dict_df = pd.DataFrame(data)
dict_df
```
