## Data Visualization with Seaborn

```python
# Plots in Seaborn
# Seaborn provides two main types of plotting functions:
"""
1. Axes level functions: These functions create a single plot on a single set of axes. Examples include `sns.scatterplot()`, `sns.lineplot()`, and `sns.barplot()`.
2. Figure level functions: These functions create a complete figure with multiple subplots.

"""

# Types of Plots in Seaborn
'''
- Scatter Plot: 
    A scatter plot is used to visualize the relationship between two continuous variables. It displays points on a Cartesian plane, where each point represents an observation in the dataset. The position of each point is determined by the values of the two variables being plotted.

- Line Plot: 
    A line plot is used to visualize the relationship between two continuous variables, where one variable is typically plotted on the x-axis and the other on the y-axis. It connects data points with a line, making it easier to identify trends and patterns in the data.

- Bar Plot: 
    A bar plot is used to visualize the relationship between a categorical variable and a continuous variable. It displays rectangular bars, where the height of each bar represents the value of the continuous variable for each category. Bar plots are useful for comparing the values of different categories and identifying patterns in the data.

- Box Plot: 
    A box plot, also known as a box-and-whisker plot, is used to visualize the distribution of a continuous variable across different categories. It displays the median, quartiles, and potential outliers in the data. The box represents the interquartile range (IQR), while the whiskers extend to the minimum and maximum values within 1.5 times the IQR.

- Violin Plot: 
    A violin plot is similar to a box plot but also includes a kernel density estimation to show the distribution of the data. It provides a more detailed view of the data distribution, allowing you to see the shape of the distribution for each category. The width of the violin plot represents the density of the data at different values, while the central line indicates the median. Violin plots are useful for comparing the distribution of a continuous variable across different categories and identifying patterns in the data.

- Heatmap: 
    A heatmap is used to visualize the relationship between two categorical variables or a categorical variable and a continuous variable. It displays data in a matrix format, where the color of each cell represents the value of the variable being plotted. Heatmaps are useful for identifying patterns and relationships in the data, such as correlations between variables or clusters of similar values.

- Pair Plot: 
    A pair plot is used to visualize the relationships between multiple continuous variables in a dataset. It creates a grid of scatter plots for each pair of variables, along with histograms or density plots on the diagonal to show the distribution of each variable. Pair plots are useful for identifying correlations and patterns between multiple variables in a dataset.

- Joint Plot: 
    A joint plot is used to visualize the relationship between two continuous variables, along with their individual distributions. It creates a scatter plot of the two variables in the center, with histograms or density plots on the top and right sides to show the distribution of each variable. Joint plots are useful for identifying correlations and patterns between two variables, as well as understanding the distribution of each variable in the context of their relationship.

- Count Plot: 
    A count plot is used to visualize the count of observations in each category of a categorical variable. It displays rectangular bars, where the height of each bar represents the count of observations for each category. Count plots are useful for comparing the frequency of different categories and identifying patterns in the data.

- Facet Plot: 
    A facet plot is used to visualize the relationship between a continuous variable and a categorical variable, while also allowing for the creation of multiple subplots based on another categorical variable. It creates a grid of plots, where each subplot represents a subset of the data based on the values of the categorical variable used for faceting. Facet plots are useful for comparing the relationship between a continuous variable and a categorical variable across different subsets of the data, and for identifying patterns and trends within those subsets.

'''
```

### Relational Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# Scatter plot -> Axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

```python
# relplot -> Figure level function -> square shaped
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

```python
# line plot
gap = px.data.gapminder()
temp_df = gap[gap['country'] == "India"]
temp_df
```

```python
# line plot -> axes level function
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# using figure level function
sns.relplot(data = temp_df, x='year', y='lifeExp', kind='line')
```

```python
gap = px.data.gapminder()
temp_df = gap[gap['country'].isin(['India', 'Brazil','Germany'])]
temp_df
```

```python
sns.relplot(kind='line',data=temp_df, x='year', y='lifeExp', hue='country', style='continent', size='continent')
```

```python
sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country')
```

```python
# face plot -> figure level function -> only work with relplot
sns.relplot(kind='scatter', data=tips, x='total_bill', y='tip', row='sex',col='smoker')
```

```python
# col wrap
sns.relplot(kind='scatter', data=gap, x='lifeExp', y='gdpPercap', col='year', col_wrap=3)
```

### Distribution Plots

```python
# figure level funciton -> displot
# axes level function -> hisplot -> kdeplot -> rugplot
```

```python
tips
```

```python
# Histogram plot
sns.histplot(data=tips, x='total_bill')
```

```python
# figure level
sns.displot(data=tips, x = 'total_bill', kind='hist')
```

```python
# bins parameter
sns.displot(data=tips, x = 'total_bill', kind='hist', bins=20)
```

```python
# Histogram on categorical data

# acting like a count plot
sns.displot(data=tips, x = 'day', kind='hist')
```

```python
sns.displot(data=tips, x = 'tip', kind='hist', hue='sex', element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic ,kind='hist', x='age', element='step', hue='sex')
```

```python
# faceting -> will not work with axes level functions
sns.displot(data=titanic ,kind='hist', x='age', col='sex', hue='survived', element='step')
```

```python
# kde plot
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True)
```

```python
# rug plot
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

```python
# Bivariat analysis 
# 2D plot in  histogram
sns.histplot(data=tips, x='total_bill', y='tip')
```

```python
sns.kdeplot(data=tips, x='total_bill', y='tip')
```
